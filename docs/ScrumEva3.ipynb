## Análisis de Ventas y Feriados

### Proyecto de Ingeniería de Datos utilizando Metodología Ágil Scrum

Este proyecto tuvo como objetivo desarrollar una solución integral de análisis de datos para una empresa minorista, permitiendo estudiar el comportamiento de las ventas y su relación con los días feriados y las metas comerciales regionales.

La solución implementada integró información proveniente de tres fuentes de datos diferentes: un archivo CSV con registros históricos de ventas, una API pública de feriados y una base de datos MySQL que almacena las metas comerciales definidas para cada región. Mediante un proceso ETL (Extract, Transform and Load), los datos fueron extraídos, limpiados, transformados e integrados para construir un Data Warehouse centralizado.

Posteriormente, la información consolidada fue utilizada para la generación de dashboards interactivos que permiten visualizar indicadores clave de negocio, analizar tendencias de ventas, evaluar el impacto de los feriados y monitorear el cumplimiento de metas comerciales.

Además, la solución fue preparada para ejecutarse en entornos reproducibles mediante Docker, facilitando su despliegue y utilización por parte de cualquier integrante del equipo.


## Problema de Negocio

La empresa minorista dispone de grandes volúmenes de información de ventas almacenados en diferentes fuentes de datos, pero actualmente no cuenta con una solución centralizada que permita integrar, analizar y visualizar dicha información de manera eficiente.

Debido a esta situación, resulta complejo evaluar el impacto que tienen los días feriados sobre el comportamiento de las ventas, así como determinar si las distintas regiones están cumpliendo con las metas comerciales establecidas por la organización.

Se requiere desarrollar una solución de Ingeniería de Datos que permita integrar información proveniente de archivos de ventas, una API de feriados y una base de datos de metas comerciales, con el fin de construir un Data Warehouse centralizado que facilite el análisis de información y la generación de dashboards para apoyar la toma de decisiones empresariales.
